```python
import random as random
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
N = 11         #Position of second absorber/cleavage

[Dpam,Da,Di,Dc] = [3,1,4,2]    #Parameters

k0 = 1         #Constante in p,q
DGmin = 0     #Delta G^-

d = 100
dinv = 1./d


#Where are the mismatches?
MM1 = [2,5]
MM2 = [8]
MM3 = [19]
MM4 = [3,8]
```

```python
Xi = np.arange(0, 1, dinv)
X = np.arange(-0.5,N-1.5,dinv)
X1 = np.zeros((d*(N-1)))
Y1 = [0]
Y2 = [0]
Y3 = [0]
Y4 = [0]

for n in range(1,N):
    D = Dpam*(n==1) + Da*(n!=1) - Di*(n in MM1) - Dc*(n==N-1)
    for x in Xi:
        Y1.append((DGmin-x*D)+ Y1[(n-1)*d])
        
for n in range(1,N):
    D = Dpam*(n==1) + Da*(n!=1) - Di*(n in MM2) - Dc*(n==N-1)
    for x in Xi:
        Y2.append((DGmin-x*D) + Y2[(n-1)*d])
        
for n in range(1,N):
    D = Dpam*(n==1) + Da*(n!=1) - Di*(n in MM3) - Dc*(n==N-1)
    for x in Xi:
        Y3.append((DGmin-x*D) + Y3[(n-1)*d])
        
for n in range(1,N):
    D = Dpam*(n==1) + Da*(n!=1) - Di*(n in MM4) - Dc*(n==N-1)
    for x in Xi:
        Y4.append((DGmin-x*D) + Y4[(n-1)*d]) 
```

```python
h = [5.5,7.5]
H = [-2,-2]
p = [1.5,.5]
P = [-3,-3]
v = [-.5,6.5]
V = [-9,-9]


suptitle = "Energy landscape"
title = '($\Delta_{PAM}, \Delta_C, \Delta_I, \Delta_{clv}, \Delta G^-$) =' + str((Dpam, Da, Di, Dc, DGmin))
plt.plot(X,X1, 'r:')
plt.plot(X,Y1[1:], color='b', label=('Mismatch at '+str(MM1)))
plt.plot(X,Y2[1:], color='g', linestyle='--', label=('Mismatch at '+str(MM2)))
plt.plot(X,Y3[1:], color='r', linestyle='-.', label=('Mismatches at '+str(MM3)))
plt.plot(X,Y4[1:], color='g', linestyle=':', label=('Mismatch at '+str(MM4)))
#plt.plot(h,H, color='black', linestyle=':')
#plt.plot(p,P, color='black', linestyle=':')
#plt.plot(v,V, color='black', linestyle=':')
plt.suptitle(suptitle)
plt.title(title)
plt.xlabel("State space")
plt.ylabel("Free energy")
plt.xticks(np.arange(0,N,1), (["PAM"] + range(2,N) + ["clv"]))
plt.yticks((),())
#plt.yticks([-9], (["Valley"]))
plt.legend()
plt.savefig('..\..\Afbeeldingen/Theory/clvub', bbox_inches = 'tight', dpi = 300)
plt.show()
```
